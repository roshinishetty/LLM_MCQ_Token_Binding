```python
# prompt: define a function to get the percentage of incorrectly predicted tokens for each file
# observation 2 results

import pandas as pd
def get_incorrect_percentage(paths):
  percentages = {}
  for path in paths:
    # Read the filtered DataFrame
    df = pd.read_csv(path)

    # Calculate the percentage of incorrectly predicted tokens
    incorrect_count = len(df[df['gold_answer'] != df['predicted_token']])
    total_count = len(df)
    percentage = incorrect_count / total_count * 100

    # Store the percentage in the dictionary
    filename = path.split('/')[-1]
    percentages[filename] = percentage

  return percentages

# Get percentages for varying_option
paths_option = ['/content/drive/MyDrive/685 Project/mmlu 02/llama-2-7b-lora-tuned-per-16k/varying_option/' + file for file in file_list]
percentages_option = get_incorrect_percentage(paths_option)

# Get percentages for varying_position
paths_position = ['/content/drive/MyDrive/685 Project/mmlu 02/llama-2-7b-lora-tuned-per-16k/varying_position/' + file for file in file_list]
percentages_position = get_incorrect_percentage(paths_position)

# Print the percentages
print("\n\nVarying Option Percentages:")
for filename, percentage in percentages_option.items():
  print(f"{filename}: {percentage:.2f}%")

print("\n\nVarying Position Percentages:")
for filename, percentage in percentages_position.items():
  print(f"{filename}: {percentage:.2f}%")
```
