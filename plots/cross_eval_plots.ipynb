```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Directory containing the LLM folders
llm_folder = '/content/drive/MyDrive/685 Project/mmlu 03'  # Update with the path to your LLM folder

# List of specific folders you want to include
specific_folders = ['gemma-2b', 'gemma-7b-it', 'gemma-2b-it', 'llama-2-7b', 'llama-2-13b', 'mistral-7b', 'mistral-7b-instruct', 'llama-3-8b', 'vicuna-7b', 'vicuna-13b', 'llama-7b', 'gemma-7b']  # Update with the names of the folders you want to include

# Get the list of LLM folders that match the specific folders
llm_folders = [os.path.join(llm_folder, folder) for folder in specific_folders if os.path.isdir(os.path.join(llm_folder, folder))]

# Initialize lists to store data
llm_names = []
data = {'0': [], '1': [], '2': [], '3': []}
means = {'0': [], '1': [], '2': [], '3': []}
std_devs = {'0': [], '1': [], '2': [], '3': []}

# Define colors
colors = plt.cm.tab10(np.linspace(0, 1, 4))  # Adjust color map for more subtle colors
colors[0] = np.array([76, 146, 195, 255])/255
colors[1] = np.array([255, 152, 62, 255])/255
colors[2] = np.array([86, 179, 86, 255])/255
colors[3] = np.array([222, 83, 83, 255])/255

# Confidence percentage and number of samples
confidence_percentage = 99  # You can change this value as needed
n = 50  # Number of samples
alpha = 1 - (confidence_percentage / 100)
z_score = norm.ppf(1 - alpha/2)

# Iterate over each LLM folder
for llm_dir in llm_folders:
    llm_name = os.path.basename(llm_dir)
    llm_names.append(llm_name)

    # Path to the recall_imbalance_ID.csv file
    csv_file = os.path.join(llm_dir, 'varying_position', 'recall_imbalance_ID.csv')

    # Read the CSV file and get the mean and standard deviation rows
    if os.path.exists(csv_file):
        df = pd.read_csv(csv_file)
        mean_row = df.iloc[-2]  # Get the second last row (mean)
        std_dev_row = df.iloc[-1]  # Get the last row (std deviation)
        for pos in ['0', '1', '2', '3']:
            mean_value = mean_row[pos]
            std_dev_value = std_dev_row[pos]
            means[pos].append(mean_value)
            std_devs[pos].append(std_dev_value)

# Create a DataFrame for the means and standard deviations
df_means = pd.DataFrame(means, index=llm_names)
df_std_devs = pd.DataFrame(std_devs, index=llm_names)

# Calculate the error margins
errors = {pos: z_score * df_std_devs[pos] / np.sqrt(n) for pos in ['0', '1', '2', '3']}

# Plotting
bar_width = 0.2
positions = np.arange(len(llm_names))
plt.figure(figsize=(12, 8))

for i, pos in enumerate(['0', '1', '2', '3']):
    plt.bar(positions + i * bar_width, df_means[pos], bar_width, yerr=errors[pos], label=f'Position {pos}', color=colors[int(pos)], capsize=5, error_kw={'ecolor': 'darkgray', 'elinewidth': 1})

plt.xlabel('LLM')
plt.ylabel('Recall Imbalance')
plt.title(f'Comparison of 0, 1, 2, 3 positions for each LLM with {confidence_percentage}% Confidence Intervals')
plt.xticks(positions + bar_width * 1.5, llm_names, rotation=45, ha='right')
plt.legend(title='Positions')
plt.tight_layout()

# Save the plot
plot_file = '/content/drive/MyDrive/685 Project/comparison_plot_recall_ID_position_with_error_bars.png'  # Save the plot in the current directory
plt.savefig(plot_file)
plt.close()

print("Plot saved as 'comparison_plot_recall_ID_position_with_error_bars.png'")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Directory containing the LLM folders
llm_folder = '/content/drive/MyDrive/685 Project/mmlu 03'  # Update with the path to your LLM folder

# List of specific folders you want to include
specific_folders = ['gemma-2b', 'gemma-2b-it','gemma-7b', 'gemma-7b-it', 'llama-7b', 'llama-2-7b', 'llama-2-13b']  # Update with the names of the folders you want to include
#'mistral-7b', 'mistral-7b-instruct',  'vicuna-7b', 'vicuna-13b', 'gemma-2b', 'gemma-2b-it', ,, 'llama-3-8b', 'vicuna-13b'

# Get the list of LLM folders that match the specific folders
llm_folders = [os.path.join(llm_folder, folder) for folder in specific_folders if os.path.isdir(os.path.join(llm_folder, folder))]

# Initialize lists to store data
llm_names = []
data = {'A': [], 'B': [], 'C': [], 'D': []}
means = {'A': [], 'B': [], 'C': [], 'D': []}
std_devs = {'A': [], 'B': [], 'C': [], 'D': []}

# Define colors
colors = plt.cm.tab10(np.linspace(0, 1, 4))  # Adjust color map for more subtle colors
colors[0] = np.array([76, 146, 195, 255])/255
colors[1] = np.array([255, 152, 62, 255])/255
colors[2] = np.array([86, 179, 86, 255])/255
colors[3] = np.array([222, 83, 83, 255])/255

# Confidence percentage and number of samples
confidence_percentage = 99  # You can change this value as needed
n = 50  # Number of samples
alpha = 1 - (confidence_percentage / 100)
z_score = norm.ppf(1 - alpha/2)

# Iterate over each LLM folder
for llm_dir in llm_folders:
    llm_name = os.path.basename(llm_dir)
    llm_names.append(llm_name)

    # Path to the recall_imbalance_ID.csv file
    csv_file = os.path.join(llm_dir, 'varying_option', 'recall_imbalance_ID.csv')

    # Read the CSV file and get the mean and standard deviation rows
    if os.path.exists(csv_file):
        df = pd.read_csv(csv_file)
        mean_row = df.iloc[-2]  # Get the second last row (mean)
        std_dev_row = df.iloc[-1]  # Get the last row (std deviation)
        for pos in ['A', 'B', 'C', 'D']:
            mean_value = mean_row[pos]*100.0
            std_dev_value = std_dev_row[pos]*100.0
            means[pos].append(mean_value)
            std_devs[pos].append(std_dev_value)

# Create a DataFrame for the means and standard deviations
df_means = pd.DataFrame(means, index=llm_names)
df_std_devs = pd.DataFrame(std_devs, index=llm_names)

# Calculate the error margins
errors = {pos: z_score * df_std_devs[pos] / np.sqrt(n) for pos in ['A', 'B', 'C', 'D']}

# Plotting
bar_width = 0.2
gap = 0.5  # Gap between groups of bars
positions = np.arange(len(llm_names)) * (4 * bar_width + gap)
plt.figure(figsize=(12, 8))

for i, pos in enumerate(['A', 'B', 'C', 'D']):
    plt.bar(positions + i * bar_width, df_means[pos], bar_width, yerr=errors[pos], label=f'Option {pos}', color=colors[i], capsize=5, error_kw={'ecolor': 'darkgray', 'elinewidth': 1})


plt.xlabel('LLM')
plt.ylabel('Recall Imbalance')
plt.title(f'Comparison of A, B, C, D options for each LLM with {confidence_percentage}% Confidence Intervals')
plt.xticks(positions + bar_width * 1.5, llm_names, rotation=45, ha='right')
plt.axhline(0, color='black', linewidth=0.8)  # Add the x-axis line back
plt.legend(title='Options')
plt.tight_layout()

# Save the plot
plot_file = '/content/drive/MyDrive/685 Project/comparison_plot_recall_ID_option_with_error_bars.png'  # Save the plot in the current directory
plt.savefig(plot_file)
plt.close()

print("Plot saved as 'comparison_plot_recall_ID_option_with_error_bars.png'")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Directory containing the LLM folders
llm_folder = '/content/drive/MyDrive/685 Project/mmlu 03'  # Update with the path to your LLM folder

# List of specific folders you want to include
specific_folders = ['gemma-2b', 'gemma-7b-it', 'gemma-2b-it', 'llama-2-7b', 'llama-2-13b', 'mistral-7b', 'mistral-7b-instruct', 'llama-3-8b', 'vicuna-7b', 'vicuna-13b', 'llama-7b', 'gemma-7b']  # Update with the names of the folders you want to include

# Get the list of LLM folders that match the specific folders
llm_folders = [os.path.join(llm_folder, folder) for folder in specific_folders if os.path.isdir(os.path.join(llm_folder, folder))]

# Initialize lists to store data
llm_names = []
data = {'top_1_bin': [], 'top_2_bin': [], 'top_3_bin': [], 'top_4_bin': []}

# Define colors
# colors = color = plt.cm.tab10(0)  # Adjust color map for more subtle colors

# Iterate over each LLM folder
for llm_dir in llm_folders:
    llm_name = os.path.basename(llm_dir)
    llm_names.append(llm_name)

    # Path to the error_bins_ID.csv file
    csv_file = os.path.join(llm_dir, 'varying_option', 'error_bins_ID.csv')

    # Read the CSV file
    if os.path.exists(csv_file):
        df = pd.read_csv(csv_file)

        # Get the last row
        last_row = df.iloc[-1]

        # Append data to the dictionary
        data['top_1_bin'].append(last_row['top_1_bin'])
        data['top_2_bin'].append(last_row['top_2_bin'])
        data['top_3_bin'].append(last_row['top_3_bin'])
        data['top_4_bin'].append(last_row['top_4_bin'])

# Create a DataFrame for the data
df_plot = pd.DataFrame(data, index=llm_names)

# Plotting
plt.figure(figsize=(10, 5))
bar_width = 0.5
positions = np.arange(len(llm_names))

bottom = np.zeros(len(llm_names))  # Bottom position for each bar

for pos in ['top_1_bin', 'top_2_bin', 'top_3_bin', 'top_4_bin']:
    plt.bar(positions, df_plot[pos], bar_width, label=pos, bottom=bottom)
    bottom += df_plot[pos]

plt.xlabel('LLM')
plt.ylabel('Percentage')
plt.title('Gold Answer in Bin Number for each LLM', pad=40)
plt.xticks(positions, llm_names, rotation=45, ha='right')
plt.legend(title='Bins', loc='upper center', bbox_to_anchor=(0.5, 1.20), ncol=4)

plt.tight_layout()

# Save the plot
plot_file = '/content/drive/MyDrive/685 Project/comparison_plot_error_bins_ID_option.png'
plt.savefig(plot_file)
plt.close()

print("Plot saved as 'comparison_plot_error_bins_ID_option.png'")
```
